```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
titanic=sns.load_dataset('titanic')
iris=sns.load_dataset('iris')
```

```python
sns.set_style('darkgrid')
```

```python
#box plot
sns.boxplot(x='alive',
           y='age',
           hue='adult_male',
           data=titanic)
```

```python
sns.boxplot(data=iris,orient='h')
```

```python
sns.violinplot(x='age',y='sex',hue='survived',data=titanic)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
sns.countplot(x='deck',data=titanic,palette='Greens_d')
```

```python
sns.pointplot(x='class',
              y='survived',
              hue='sex',
              data=titanic,
              palette={'male':'g','female':'m'},
              markers=['^','o'],
              linestyles=['-','--'])
```

```python
sns.stripplot(x='species',
              y='petal_length',
              data=iris)
```

```python
sns.swarmplot(x='species',
            y='petal_length',
             data=iris)
```

```python
g=sns.FacetGrid(titanic,col='survived',row='sex')
g=g.map(plt.hist,'age')
```

```python
#Instead of 2 steps FacetGrid, I can use catplot or lmplot instead
sns.set_style('white')
df=pd.read_csv('college_data.csv')
sns.catplot(data=df,
        x='SAT_AVG_ALL',
        kind='point',
        row='Degree_Type',
        row_order=['Graduate', 'Bachelors', 'Associates', 'Certificate'],
        height=2,aspect=5
           )

plt.show()
plt.clf()
```

```python
g=sns.FacetGrid(df,col='HIGHDEG')
g.map(sns.boxplot,'Tuition',order=['1','2','3','4'])
```

```python
sns.catplot(x='Tuition',data=df,col='HIGHDEG',kind='box')
```

```python
sns.lmplot(data=df,x='Tuition',y='SAT_AVG_ALL',col='HIGHDEG',fit_reg=False)
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot('sepal_length','sepal_width',data=iris,kind='kde')
```
